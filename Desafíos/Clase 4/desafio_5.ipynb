**1.** Guardar en otro archivo, en formato json, las canciones que tienen asignado más de un género.

---

- Importo los módulos necesarios y creo la ruta al archivo.

```python
from pathlib import Path
import csv
import json

file_route = Path('Archivos_Ejemplos') / "songs_normalize.csv"
```

- Abro el archivo y lo leo con csv.DictReader para obtener diccionarios con los datos que voy a poner en una lista (data) para concordar con el formato json.

```python
with open(file_route, "r", encoding= 'utf-8') as spotify_file:
    csv_reader = csv.DictReader(spotify_file, delimiter= ',')
    data = list(csv_reader)

data
```

- Antes de crear el nuevo archivo filtro las canciones que me interesan (aquellas con más de un género).
- Como el método .split() divide mis strings por espacios creando listas de subcadenas, puedo utilizar una función lambda que chequee si la longitud de cada canción en el indice de sus generos tiene 2 o más elementos cuando le aplico .split().

```python
new_data = list(filter(lambda song: len(song['genre'].split()) >= 2, data))

new_data
```

- Finalmente creo el archivo json y serializo los datos.

```python
new_file_route = Path('Archivos_Ejemplos') / "songs_normalize.json"

with open(new_file_route, "w") as new_spotify_file:
    json.dump(new_data, new_spotify_file, indent= 4)
```

**2.** Los cinco (5) artistas con más canciones en el dataset durante el año 2019.

---

- Voy a aprovechar que ya tengo los datos cargados en data y primero voy a filtrar por año.

```python
data_2019 = list(filter(lambda song: song['year'] == '2019', data))

data_2019
```

- Con el módulo Counter voy a contar cuales son los 5 artistas con más apariciones.
- Imprimo los datos con el método enumerate que busqué en internet y me ayudó a representar las posiciones del TOP.

```python
from collections import Counter

artist_counter = Counter(song['artist'] for song in data_2019)
most_aparisons_artist = artist_counter.most_common(5)

print("TOP 5 ARTISTAS CON MÁS APARICIONES (AÑO 2019):")
for i, (artist, aparisons) in enumerate(most_aparisons_artist, start=1):
    print(f"{i}. {artist}")
```
